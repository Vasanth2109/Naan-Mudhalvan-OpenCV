```python
import torch

# Load the custom or Roboflow model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/best.pt')  # or a Roboflow-exported path

# Set confidence and IOU thresholds lower
model.conf = 0.25  # default is 0.25, lower it if some weapons aren't detected
model.iou = 0.45   # default is 0.45, can lower a bit to avoid merging boxes
model.max_det = 100  # increase max number of detections

# Inference
results = model('/content/Untitled6.jpg')

# Show results
results.print()
results.show()
```
